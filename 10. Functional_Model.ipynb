```python
# Lets suppose we have three inputs column say salary , height and maritial status of some people and
# we have to predict what ia age of the person and what is age and which city does he live (delhi/mumbai)
# So , basically we have two output so we'll use functional model
```

```python
from keras.layers import *
# Input layer : input shape is 3 since we had 3 columns
x = Input(shape=(3,))
# First hidden layer is getting input from x
hidden1 = Dense(128 , activation='relu')(x)
# second hidden layer is getting input from hidden1
hidden2 = Dense(64 , activation='relu')(hidden1)

# Output layers : 1-To calculate age(activation will be linear since regression)
#                 2-To calculate city(activation is sigmoid since classification)
output1 = Dense(1 , activation='linear')(hidden2)
output2 = Dense(1 , activation='sigmoid')(hidden2)
```

```python
from keras.models import Model
model = Model(inputs = x , outputs = [output1 , output2])
# Now we we make object of model , we have to mention which layer of our network is input layer and which is output layer
```

```python
model.summary()
```

```python
# Visulaization of model architecture
from keras.utils import plot_model
plot_model(model , show_shapes=True)
```
